```python
df =pd.read_csv('/content/diabetes.csv')
```

```python
# Plot a histogram for the 'Age' feature
plt.hist(df['Age'], bins=20, color='skyblue', edgecolor='black')
plt.title('Age Distribution')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```

```python
# Scatter plot between 'Age' and 'Glucose'
plt.scatter(df['Age'], df['Glucose'], alpha=0.6)
plt.title('Age vs Glucose')
plt.xlabel('Age')
plt.ylabel('Glucose')
plt.show()
```

```python
# Line plot for 'BMI' vs 'Blood Pressure'
plt.plot(df['BMI'], df['BloodPressure'], color='green', marker='o', linestyle='dashed')
plt.title('BMI vs Blood Pressure')
plt.xlabel('BMI')
plt.ylabel('Blood Pressure')
plt.show()
```

```python
# Pair plot to see relationships between several features

sns.pairplot(df[['Age', 'BMI', 'Insulin', 'Glucose', 'BloodPressure']])
plt.show()
```

```python
# Calculate correlation matrix
corr_matrix = df.corr()

# Heatmap of correlations
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt='.2f', linewidths=0.5)
plt.title('Correlation Heatmap')
plt.show()
```

```python
# If you have a target column that indicates diabetes outcome (e.g., 1 for diabetes, 0 for no diabetes)
target_count = df['Outcome'].value_counts()

# Pie chart of target values (diabetes vs non-diabetes)
plt.pie(target_count, labels=['No Diabetes', 'Diabetes'], autopct='%1.1f%%', colors=['lightcoral', 'lightgreen'])
plt.title('Diabetes Outcome Distribution')
plt.show()
```

```python
plt.scatter(df['Age'], df['Glucose'], alpha=0.6, color='purple')
plt.title('Age vs Glucose Level', fontsize=14, fontweight='bold')
plt.xlabel('Age', fontsize=12)
plt.ylabel('Glucose', fontsize=12)
plt.grid(True)
plt.show()
```

```python
# Violin plot for 'Age' vs 'BMI'
sns.violinplot(x='Age', y='BMI', data=df, palette='muted')
plt.title('Violin Plot of Age vs BMI')
plt.show()
```

```python
# Box plot for 'BMI' feature
plt.boxplot(df['BMI'])
plt.title('BMI Distribution')
plt.ylabel('BMI')
plt.show()
```

```python
z = df['Age']
x = df['BMI']
y = df['Glucose']
fig = plt.figure()
ax = plt.axes(projection='3d')
sc = ax.scatter(x, y, z, c=z, cmap='Blues')
ax.set_xlabel('BMI')
ax.set_ylabel('Glucose')
ax.set_zlabel('Age')
plt.colorbar(sc)
plt.show()
```

```python
fig2 = plt.figure()
ax = fig2.add_axes([0,0,1,1])
ax.scatter(x,y,color='g')
ax.scatter(x,z,color='b')
plt.show()
```

```python
features = ['BMI', 'Glucose', 'Insulin', 'BloodPressure', 'Age']
average_values = df[features].mean()
average_values.plot(kind='bar', color='skyblue', edgecolor='black')
plt.title('Average Feature Values in Diabetes Dataset')
plt.xlabel('Features')
plt.ylabel('Average Value')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
